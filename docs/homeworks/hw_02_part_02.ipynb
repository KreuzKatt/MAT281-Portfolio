Primero se solucionar√° para cada columna el problema de los valores vacios segun el siguiente criterio:


*   Age- Promedio de Age

*   Cabin- "-"

*   Fare- Promedio de Fare

*   Embarked- "-"

Al rellenar los datos de esta forma no perdemos el hecho de que faltan datos y a la vez no entorpecen el analisis de problema.

````python
vacios = ['Age', 'Fare', 'Cabin', 'Embarked']

vacios_dic = {'Age' : round(df_train['Age'].mean(),1), 'Fare' : round(df_train['Fare'].mean(),1), 'Cabin' : '-', 'Embarked': '-'}


for col in vacios:
  df_train[col] = df_train[col].fillna(vacios_dic[col])
  df_test[col] = df_test[col].fillna(vacios_dic[col])
````

## Transformacion de Datos

###Variables categoricas a variables numericas

**NOTA IMPORTATE:** Apartir de este punto las variables terminadas en "te" se asocian a los datos de testeo.

````python
#Separamos el DataFrame en las VAriables CAtegoricas y num√©ricas que VAYAMOS A UTILIZAR
#(Se descartar√°n las que no)
cat_val = ['Sex', 'Embarked']
num_val = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
numte_val = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
df_num = df_train[num_val]
df_cat = df_train[cat_val]
df_numte = df_test[numte_val]
df_catte = df_test[cat_val]
df_cat.head()
````

````python
#Datos de Entrenamiento
df_cat2 = df_cat
print('Metodo de Clasificacion:')

for col in df_cat.columns:
  print(f'\n Para la columna {col}:')
  ndf_cat = pd.get_dummies(df_cat[col])
  i = 0
  for subcol in ndf_cat.columns:
    print("{:<30} \t = {:<5}".format(subcol, i))
    i += 1

  for fil in range(len(df_cat)):
    df_cat2.loc[fil, col]= list(ndf_cat.iloc[fil]).index(1)

#Datos de Testeo
df_catte2 = df_catte

for col in df_catte.columns:
  ndf_catte = pd.get_dummies(df_catte[col])

  for fil in range(len(df_catte)):
    df_catte2.loc[fil, col]= list(ndf_catte.iloc[fil]).index(1)
````

````python
#Dataframe de las variables categoricas pasado a variables numericas bajo el criterio impreso anteriormente
df_cat.head()
````

````python
#concatenamos dataframes
df_procesado = pd.concat([df_num, df_cat], axis=1)
df_procesadote = pd.concat([df_numte, df_catte], axis=1)
df_procesado.head()
````

````python
df_procesadote.head()
````

###Normalizar

````python
#normalizamiento
scaler = MinMaxScaler()
scaler.fit(df_procesado.drop('Survived', axis =1))

#transformamos para variables numericas
new_df = scaler.transform(df_procesado.drop('Survived', axis =1))
newte_df = scaler.transform(df_procesadote)
scaled_df = pd.DataFrame(new_df, columns= df_procesado.drop('Survived', axis =1).columns)
scaledte_df = pd.DataFrame(newte_df, columns= df_procesadote.columns)
````

````python
scaled_df.head()
````

````python
scaledte_df.head()
````

#Selecci√≥n de modelo

Se seleccionar√°n 4 m√©todos diferentes:
*   Random Forest (rf)
*   Support Vector Machine (svr)
*   Gradient Boosting (GBR)
*   Regresi√≥n Lineal (linear)

luego a partir de los resulatados se promediaran las salidas de cada m√©todo y se clasificar√° en si la persona sovrevivi√≥ o no.

````python
#separamos los datos a utilizar
X_train_var = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex',
       'Embarked']

X_train = scaled_df[X_train_var]

y_train = df_train['Survived']

X_test = scaledte_df
````

````python
#modelos
linear_model = LinearRegression() #Modelo de Regresi√≥n Lineal
svr_model = SVR() #Modelo de Support Vector Machine
rf_model = RandomForestRegressor() #Modelo de Random Forest
GBR_model = GradientBoostingRegressor() #Modelo Gradient Boosting
````

##Random Forest

````python
start_time = time.time()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_time = round(time.time() - start_time,3)
print(f'Tiempo demorado: {rf_time} segundos')
````

## Support Vector Machine

````python
start_time = time.time()
svr_model.fit(X_train, y_train)
svr_pred = svr_model.predict(X_test)
svr_time = round(time.time() - start_time,3)
print(f'Tiempo demorado: {svr_time} segundos')
````

## Gradient Boosting

````python
start_time = time.time()
GBR_model.fit(X_train, y_train)
GBR_pred = GBR_model.predict(X_test)
GBR_time = round(time.time() - start_time,3)
print(f'Tiempo demorado: {GBR_time} segundos')
````

##Regresi√≥n Lineal

````python
start_time = time.time()
linear_model.fit(X_train, y_train)
linear_pred = linear_model.predict(X_test)
linear_time = round(time.time() - start_time,3)
print(f'Tiempo demorado: {linear_time} segundos')
````

#Visualizaciones del modelo

````python
rf_predf = [0]
svr_predf = [0]
GBR_predf = [0]
linear_predf = [0]

for i in range(len(rf_pred)):
  rf_predf.append(rf_pred[i])
  svr_predf.append(svr_pred[i])
  GBR_predf.append(GBR_pred[i])
  linear_predf.append(linear_pred[i])

rf_predf.remove(0)
svr_predf.remove(0)
GBR_predf.remove(0)
linear_predf.remove(0)

result_df = pd.DataFrame()
result_df['PassengerId'] = df_test['PassengerId']
result_df['Survived_rf'] = rf_predf
result_df['Survived_svr'] = svr_predf
result_df['Survived_gbr'] = GBR_predf
result_df['Survived_lin'] = linear_predf
result_df.head()
````

#An√°lisis de resultados

Individualmente cada Modelo tardo menos de 10 segundos procesando los datos, por lo que considerando esto se puede decir que trabajaron de manera eficiente y los datos entregados fueron valores entre 0 y 1.

````python
seguridad = [0]
for fil in range(result_df.shape[0]):
  temp = round((result_df['Survived_gbr'][fil] + result_df['Survived_svr'][fil] + result_df['Survived_rf'][fil] + result_df['Survived_lin'][fil])/4)
  seguridad.append(temp)
seguridad.remove(0)
result_df['Survived'] = seguridad
result_df.head()
````

````python
plot = result_df.Survived.value_counts()
plot.plot(kind="bar", title = 'Sobrevivientes')
````

A partir de los datos de testeo se concluye que alrededor de dos tercios de las personas fallecieron en el accidente.
