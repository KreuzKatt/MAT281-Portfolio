```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_04/data/company_sales_data.csv")
df.head(12)
```

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
fig = plt.figure(figsize=(10, 5))
plt.title("Ganacias a través de los meses") # agregar titulo
plt.xlabel('Mes') # nombre eje x
plt.ylabel('Total Profit') # nombre eje y
plt.grid(True)
sns.lineplot(
        x='month_number',
        y='total_profit',
        data=df
)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
plt.title("Ganacias a través de los meses") # agregar titulo
plt.xlabel('Mes') # nombre eje x
plt.ylabel('Total Profit') # nombre eje y
plt.grid(True)
sns.scatterplot(
        x='month_number',
        y='total_profit',
        data=df
)
plt.show()

```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
fig = plt.figure(figsize=(7, 5))
plt.title("Ventas por Productos") # agregar titulo
plt.xlabel('Mes') # nombre eje x
plt.ylabel('Ventas') # nombre eje y
palette = sns.color_palette("hls", 6) #Escoger colores

i = 0;
coleccion = []
for datos in df.columns[1:7]: #Agregar cada linea de productos
    sns.lineplot(
        x='month_number',
        y= datos,
        data=df,
        color = palette[i],
        marker='o'

    )
    coleccion.append(mpatches.Patch(color=palette[i], label=datos))
    i = i+1

plt.legend(handles=coleccion, fontsize="8", title= "Productos")
plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
df[['facecream', 'facewash']].plot.bar()
plt.title('Ventas por Producto')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
fig = plt.figure(figsize=(9, 6))
dfcopia=df.drop(['month_number','total_units','total_profit'], axis=1)
sns.boxplot(data=dfcopia)

plt.title('Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Ventas')
#plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
fig = plt.figure(figsize=(6, 6))
nombres = df.columns[1:7]
cantidades = []
for listas in df.columns[1:7]:
    cantidades.append( df[listas].sum())

plt.title("Productos Vendidos durante el año\n")
plt.pie(cantidades, labels=nombres, autopct="%0.1f %%")
plt.axis("equal")
plt.show()
```
